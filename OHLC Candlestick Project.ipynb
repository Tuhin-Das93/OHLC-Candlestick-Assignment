## Project 1: Understanding the Candlestick trend pattern of Stock Market Incides

### Visualizing the share price of Nifty and NASDAQ before the onset of COVID-19 in the year 2020 till post COVID era (i.e., 16th March 2021)

```python
import datetime as dt
import pandas_datareader as web
import matplotlib.pyplot as plt
```

```python
import matplotlib.dates as mdates
```

```python
from mplfinance.original_flavor import candlestick_ohlc
```

```python
start = dt.datetime(2020,1,1)
end = dt.datetime.now()
```

```python
ticker ="^NSEI" ## Nifty Data
ticker1 = "^IXIC" ## NASDAQ Composite Data

data = web.DataReader(ticker, "yahoo", start, end)
data1 = web.DataReader(ticker1, "yahoo", start, end)
```

```python
data = data[["Open", "High", "Low", "Close"]]
data1 = data1[["Open", "High", "Low", "Close"]]
```

```python
data.reset_index(inplace=True)
data1.reset_index(inplace=True)
```

```python
data["Date"] = data["Date"].map(mdates.date2num)
data1["Date"] = data1["Date"].map(mdates.date2num)
```

```python
plt.figure(figsize=(30,10))
ax = plt.subplot()
ax.grid(True)
ax.set_title("{} Share Price".format(ticker), color="white",fontdict={'fontsize': 25, 'fontweight': 'medium'})

ax.set_axisbelow(True)
ax.set_facecolor("black")
ax.figure.set_facecolor("#121212")
ax.tick_params(axis="x", colors="white",labelsize=20 )
ax.tick_params(axis="y", colors="white", labelsize=20)
ax.xaxis_date()

candlestick_ohlc(ax, data.values, width=0.5, colorup="#00ff00")

plt.show()
```

```python
plt.figure(figsize=(30,10))
ax = plt.subplot()
ax.grid(True)
ax.set_axisbelow(True)
ax.set_title("{} Share Price".format(ticker1), color="white",fontdict={'fontsize': 25, 'fontweight': 'medium'})
ax.set_facecolor("black")
ax.figure.set_facecolor("#121212")
ax.tick_params(axis="x", colors="white",labelsize=20 )
ax.tick_params(axis="y", colors="white", labelsize=20)
ax.xaxis_date()

candlestick_ohlc(ax, data1.values, width=0.5, colorup="#00ff00")

plt.show()
```
